```python
def analyze_column(col):
    categories = df[col].unique()
    count = df[col].value_counts()
    ucount = df[col].nunique()
 # print(UAA_SCI[col].sample(10))
    print("\n\n\n ------",col,"------")
    if pd.api.types.is_numeric_dtype(df[col]) == True:
        print("\nNumeric DataType")
    elif pd.api.types.is_string_dtype(df[col]) == True:
        print("\nString DataType")
    else:
        print("Unknown Datatype")
    #print("\n\n\nCategories:",categories)
    print("\nNumber of Categories:",ucount)
    print("\nCounts:\n",count)
```

```python
for column in df.columns:
    analyze_column(column)
```
